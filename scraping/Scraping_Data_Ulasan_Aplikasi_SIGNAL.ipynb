## **Scraping Data Aplikasi SIGNAL Di Google Play Store**

Menggunakan tools ```google-play-scraper``` untuk mengambil data ulasan pengguna.

````python
!pip install google-play-scraper
````

````python
# Import google-play-scraper
from google_play_scraper import reviews_all, reviews
````

````python
# Mengambil ulasan pengguna aplikasi SIGNAL
results = reviews(
    'app.signal.id',
    lang='id',
    country='id',
    count=15000
)
````

````python
# Preview hasil scraping
results[0]
````

````python
# Cek jumlah data yang telah di-scrape
len(results[0])
````

## **Mengubah Data Scrape ke Dalam Dataframe**

Menggunakna library ```pandas DataFrame``` untuk mengubah data hasil scrape ke dalam format dataframe.

````python
# Simpan data hasil scrape ke dalam dataframe
import pandas as pd
df = pd.DataFrame(results[0])
df
````

## **Manipulasi Dataframe (Hanya Ambil Kolom Ulasannya)**

````python
# Buat salinan dataframe scrape
df_copy = df.copy()
````

````python
# Cek informasi salinan
df_copy.info()
````

````python
# Simpan hanya kolom 'reviewID' dan 'content' yang berisi ulasan pengguna
signal_reviews = df_copy[['reviewId', 'content']]
signal_reviews
````

````python
# Simpan dataframe signal ke format .csv
signal_reviews.to_csv('signal_reviews.csv', index=False)
````
